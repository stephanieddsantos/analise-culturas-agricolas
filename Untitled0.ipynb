```python
# Agrupamento de culturas por requisitos similares
# Padronizando os dados para a análise de cluster
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# Selecionando apenas as colunas numéricas para clustering
X = df_pt[['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']]

# Padronizando os dados
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Aplicando PCA para visualização em 2D
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Adicionando os componentes principais ao dataframe
df_pt['PC1'] = X_pca[:, 0]
df_pt['PC2'] = X_pca[:, 1]

# Determinando o número ideal de clusters usando o método do cotovelo
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

# Plotando o gráfico do cotovelo
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='-')
plt.title('Método do Cotovelo para Determinação do Número Ideal de Clusters')
plt.xlabel('Número de Clusters')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()

# Com base no gráfico do cotovelo, escolhemos um número adequado de clusters (por exemplo, 4)
n_clusters = 4
kmeans = KMeans(n_clusters=n_clusters, init='k-means++', max_iter=300, n_init=10, random_state=42)
df_pt['cluster'] = kmeans.fit_predict(X_scaled)

# Visualizando os clusters no espaço PCA
plt.figure(figsize=(12, 8))
for cluster in range(n_clusters):
    plt.scatter(
        df_pt[df_pt['cluster'] == cluster]['PC1'],
        df_pt[df_pt['cluster'] == cluster]['PC2'],
        label=f'Cluster {cluster+1}'
    )

plt.title('Agrupamento de Culturas por Requisitos Similares')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.legend()
plt.grid(True)
plt.show()

# Analisando quais culturas estão em cada cluster
for cluster in range(n_clusters):
    print(f"\nCulturas no Cluster {cluster+1}:")
    cluster_cultures = df_pt[df_pt['cluster'] == cluster]['cultura'].unique()
    for culture in cluster_cultures:
        print(f"- {culture}")
```

```python
# Função para recomendar culturas com base nas condições fornecidas
def recomendar_culturas(n, p, k, temperatura, umidade, ph, precipitacao):
    # Criando um dataframe com as condições fornecidas
    condicoes = pd.DataFrame({
        'N': [n],
        'P': [p],
        'K': [k],
        'temperature': [temperatura],
        'humidity': [umidade],
        'ph': [ph],
        'rainfall': [precipitacao]
    })

    # Padronizando as condições usando o mesmo scaler
    condicoes_scaled = scaler.transform(condicoes)

    # Calculando a distância euclidiana entre as condições fornecidas e cada cultura
    distancias = []
    for cultura in df_pt['cultura'].unique():
        # Obtendo os valores médios para a cultura
        valores_medios = df_pt[df_pt['cultura'] == cultura][['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']].mean().values

        # Padronizando os valores médios
        valores_medios_scaled = scaler.transform([valores_medios])[0]

        # Calculando a distância euclidiana
        distancia = np.sqrt(np.sum((condicoes_scaled[0] - valores_medios_scaled) ** 2))

        distancias.append((cultura, distancia))

    # Ordenando as culturas pela distância (menor para maior)
    distancias_ordenadas = sorted(distancias, key=lambda x: x[1])

    return distancias_ordenadas

# Exemplo de uso da função de recomendação
# Vamos usar valores médios para demonstração
n_medio = df_pt['N'].mean()
p_medio = df_pt['P'].mean()
k_medio = df_pt['K'].mean()
temp_media = df_pt['temperature'].mean()
umidade_media = df_pt['humidity'].mean()
ph_medio = df_pt['ph'].mean()
precipitacao_media = df_pt['rainfall'].mean()

print("Valores médios utilizados para demonstração:")
print(f"N: {n_medio:.2f}, P: {p_medio:.2f}, K: {k_medio:.2f}")
print(f"Temperatura: {temp_media:.2f}°C, Umidade: {umidade_media:.2f}%")
print(f"pH: {ph_medio:.2f}, Precipitação: {precipitacao_media:.2f} mm")

# Obtendo as recomendações
recomendacoes = recomendar_culturas(n_medio, p_medio, k_medio, temp_media, umidade_media, ph_medio, precipitacao_media)

print("\nCulturas recomendadas (da mais adequada para a menos adequada):")
for i, (cultura, distancia) in enumerate(recomendacoes[:10]):
    print(f"{i+1}. {cultura} - Distância: {distancia:.4f}")
```

```python
# Interface interativa para recomendação de culturas
from ipywidgets import interact, widgets
import numpy as np

# Função para recomendar culturas com base nas condições fornecidas
def recomendar_culturas(n, p, k, temperatura, umidade, ph, precipitacao):
    # Criando um dataframe com as condições fornecidas
    condicoes = pd.DataFrame({
        'N': [n],
        'P': [p],
        'K': [k],
        'temperature': [temperatura],
        'humidity': [umidade],
        'ph': [ph],
        'rainfall': [precipitacao]
    })

    # Padronizando as condições usando o mesmo scaler
    condicoes_scaled = scaler.transform(condicoes)

    # Calculando a distância euclidiana entre as condições fornecidas e cada cultura
    distancias = []
    for cultura in df_pt['cultura'].unique():
        # Obtendo os valores médios para a cultura
        valores_medios = df_pt[df_pt['cultura'] == cultura][['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']].mean().values

        # Padronizando os valores médios
        valores_medios_scaled = scaler.transform([valores_medios])[0]

        # Calculando a distância euclidiana
        distancia = np.sqrt(np.sum((condicoes_scaled[0] - valores_medios_scaled) ** 2))

        distancias.append((cultura, distancia))

    # Ordenando as culturas pela distância (menor para maior)
    distancias_ordenadas = sorted(distancias, key=lambda x: x[1])

    return distancias_ordenadas

@interact(
    n=widgets.IntSlider(min=0, max=140, step=5, value=50, description='Nitrogênio (N):'),
    p=widgets.IntSlider(min=5, max=145, step=5, value=50, description='Fósforo (P):'),
    k=widgets.IntSlider(min=5, max=205, step=5, value=50, description='Potássio (K):'),
    temperatura=widgets.FloatSlider(min=8, max=44, step=0.5, value=25, description='Temperatura (°C):'),
    umidade=widgets.FloatSlider(min=14, max=100, step=1, value=70, description='Umidade (%):'),
    ph=widgets.FloatSlider(min=3.5, max=10, step=0.1, value=6.5, description='pH:'),
    precipitacao=widgets.FloatSlider(min=20, max=300, step=5, value=100, description='Precipitação (mm):')
)
def recomendar_interativo(n, p, k, temperatura, umidade, ph, precipitacao):
    recomendacoes = recomendar_culturas(n, p, k, temperatura, umidade, ph, precipitacao)
    print("Top 5 culturas recomendadas para as condições selecionadas:")
    for i, (cultura, distancia) in enumerate(recomendacoes[:5]):
        print(f"{i+1}. {cultura} - Compatibilidade: {100 * (1 - distancia/10):.1f}%")

    # Visualizando as características das culturas recomendadas
    culturas_recomendadas = [cultura for cultura, _ in recomendacoes[:5]]
    visualizar_caracteristicas(culturas_recomendadas)

# Função para visualizar características das culturas selecionadas
def visualizar_caracteristicas(culturas):
    # Criando o gráfico de radar
    fig, axes = plt.subplots(2, 3, figsize=(15, 10), subplot_kw={'polar': True})
    axes = axes.flatten()

    # Colunas para o gráfico radar
    colunas = ['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']

    # Número de variáveis
    N = len(colunas)

    # Calculando o ângulo para cada eixo
    angulos = [n / float(N) * 2 * np.pi for n in range(N)]
    angulos += angulos[:1]  # Fechando o gráfico

    # Para cada cultura, criar um gráfico radar
    for i, cultura in enumerate(culturas):
        if i < len(axes):  # Garantir que não exceda o número de subplots
            ax = axes[i]

            # Obtendo os valores médios para a cultura atual
            valores = []
            for coluna in colunas:
                valor_medio = df_pt[df_pt['cultura'] == cultura][coluna].mean()
                valores.append(valor_medio)

            # Fechando o gráfico
            valores += valores[:1]

            # Desenhando o gráfico
            ax.plot(angulos, valores, linewidth=2, linestyle='solid')
            ax.fill(angulos, valores, alpha=0.25)

            # Adicionando os rótulos
            ax.set_xticks(angulos[:-1])
            ax.set_xticklabels(colunas, size=8)

            # Adicionando o título
            ax.set_title(cultura, size=12)

    # Removendo subplots extras, se houver
    for i in range(len(culturas), len(axes)):
        fig.delaxes(axes[i])

    plt.tight_layout()
    plt.show()
```

```python
# Interface interativa para recomendação de culturas - versão simplificada
from ipywidgets import interact, widgets
import numpy as np

# Função para recomendar culturas com base nas condições fornecidas
def recomendar_culturas(n, p, k, temperatura, umidade, ph, precipitacao):
    # Criando um dataframe com as condições fornecidas
    condicoes = pd.DataFrame({
        'N': [n],
        'P': [p],
        'K': [k],
        'temperature': [temperatura],
        'humidity': [umidade],
        'ph': [ph],
        'rainfall': [precipitacao]
    })

    # Padronizando as condições usando o mesmo scaler
    condicoes_scaled = scaler.transform(condicoes)

    # Calculando a distância euclidiana entre as condições fornecidas e cada cultura
    distancias = []
    for cultura in df_pt['cultura'].unique():
        # Obtendo os valores médios para a cultura
        valores_medios = df_pt[df_pt['cultura'] == cultura][['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']].mean().values

        # Padronizando os valores médios
        valores_medios_scaled = scaler.transform([valores_medios])[0]

        # Calculando a distância euclidiana
        distancia = np.sqrt(np.sum((condicoes_scaled[0] - valores_medios_scaled) ** 2))

        distancias.append((cultura, distancia))

    # Ordenando as culturas pela distância (menor para maior)
    distancias_ordenadas = sorted(distancias, key=lambda x: x[1])

    return distancias_ordenadas

@interact(
    n=widgets.IntSlider(min=0, max=140, step=5, value=50, description='Nitrogênio (N):'),
    p=widgets.IntSlider(min=5, max=145, step=5, value=50, description='Fósforo (P):'),
    k=widgets.IntSlider(min=5, max=205, step=5, value=50, description='Potássio (K):'),
    temperatura=widgets.FloatSlider(min=8, max=44, step=0.5, value=25, description='Temperatura (°C):'),
    umidade=widgets.FloatSlider(min=14, max=100, step=1, value=70, description='Umidade (%):'),
    ph=widgets.FloatSlider(min=3.5, max=10, step=0.1, value=6.5, description='pH:'),
    precipitacao=widgets.FloatSlider(min=20, max=300, step=5, value=100, description='Precipitação (mm):')
)
def recomendar_interativo(n, p, k, temperatura, umidade, ph, precipitacao):
    recomendacoes = recomendar_culturas(n, p, k, temperatura, umidade, ph, precipitacao)
    print("Top 5 culturas recomendadas para as condições selecionadas:")
    for i, (cultura, distancia) in enumerate(recomendacoes[:5]):
        print(f"{i+1}. {cultura} - Compatibilidade: {100 * (1 - distancia/10):.1f}%")

    # Visualizando as características das culturas recomendadas
    culturas_top5 = [cultura for cultura, _ in recomendacoes[:5]]

    # Criando um dataframe com os valores médios para cada cultura recomendada
    dados_radar = []
    for cultura in culturas_top5:
        # Filtrando o dataframe para a cultura atual
        df_cultura = df_pt[df_pt['cultura'] == cultura]

        # Calculando os valores médios para cada parâmetro
        n_medio = df_cultura['N'].mean()
        p_medio = df_cultura['P'].mean()
        k_medio = df_cultura['K'].mean()
        temp_media = df_cultura['temperature'].mean()
        umidade_media = df_cultura['humidity'].mean()
        ph_medio = df_cultura['ph'].mean()
        precip_media = df_cultura['rainfall'].mean()

        # Adicionando os valores ao dataframe
        dados_radar.append({
            'Cultura': cultura,
            'N': n_medio,
            'P': p_medio,
            'K': k_medio,
            'Temperatura': temp_media,
            'Umidade': umidade_media,
            'pH': ph_medio,
            'Precipitação': precip_media
        })

    # Convertendo a lista para um dataframe
    df_radar = pd.DataFrame(dados_radar)

    # Exibindo os valores médios para cada cultura
    print("\nValores médios para as culturas recomendadas:")
    print(df_radar.set_index('Cultura'))

    # Criando um gráfico de barras para comparar os parâmetros
    plt.figure(figsize=(15, 10))

    # Parâmetros a serem plotados
    parametros = ['N', 'P', 'K', 'Temperatura', 'Umidade', 'pH', 'Precipitação']

    # Criando um subplot para cada parâmetro
    for i, param in enumerate(parametros):
        plt.subplot(3, 3, i+1)
        sns.barplot(x='Cultura', y=param, data=df_radar)
        plt.title(f'{param} por Cultura')
        plt.xticks(rotation=45)

    plt.tight_layout()
    plt.show()
```
